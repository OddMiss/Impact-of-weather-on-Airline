```python
# Beijing time
begin_date = pd.to_datetime('2024-4-19')
end_date = pd.to_datetime('2024-5-19')
date_index = begin_date
origin_list = []
while date_index <= end_date:
    month = date_index.month
    day = date_index.day
    file_out = f"{month}.{day}出港.csv"
    df = pd.read_csv(path_out + file_out,
                     parse_dates=['flightDate'])
    origin_list.append(df)
    date_index += timedelta(days=1)
origin_df = pd.concat(origin_list).reset_index(drop=True)
origin_df
```

```python
origin_df.to_csv('Airport.csv', index=False)
```
