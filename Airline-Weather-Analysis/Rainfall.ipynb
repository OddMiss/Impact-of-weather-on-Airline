```python
# Extract "11" data from each daily entry and create a list of tuples
data_list = []
for entry in json_data["value"][0]["responses"]:
    for date_key, values in entry["trendChart"].items():
        if "11" in values["trendDays"]:
            data_list.append((date_key, values["trendDays"]["11"]))

# Convert the list of tuples into a DataFrame
Rainfall = pd.DataFrame(data_list, columns=["Date", "Rainfall/mm"])
Rainfall
```
