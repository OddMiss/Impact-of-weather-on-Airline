```python
import pandas as pd
import numpy as np
from datetime import timedelta

path_origin = "./Weather/"
path_api = "./Weather_api_data/"
month = 4
day = 1
hour = 0
```

Beijing Time = UTC time + 8h

Train date duration: 4.19-5.19

```python
def Generate_Origin(month, day, hour, path):
    '''
    Function of generating origin dataframe
    Attention: Input date is UTC time.
    Beijing time = UTC time + 8h.
    '''
    file = 'weather_data_2024_{}_{}_{}.xlsx'.format(month, day, hour)
    origin = pd.read_excel(path + file)
    
    # Transfer to normal dataframe
    df = pd.DataFrame({origin[0][i]: [origin[1][i]] for i in range(len(origin[0]))})
    
    # UTC to Beijing time
    date = pd.to_datetime(f"2024-{month}-{day} {hour}:00") + timedelta(hours=8)
    # Add a new column named "Date"
    date_data = {"Date": [date]}
    # Create a DataFrame for the date data
    date_df = pd.DataFrame(date_data)
    
    # Concatenate the date column with the existing DataFrame
    df = pd.concat([date_df, df], axis=1)
    # Convert the "Date" column to datetime format
    # df['Date'] = pd.to_datetime(df['Date'])
    return df
```

```python
# UTC time
begin_date = pd.to_datetime('2024-4-19 00:00') - timedelta(hours=8)
end_date = pd.to_datetime('2024-5-20 00:00') - timedelta(hours=8)
print(begin_date.month)
print(begin_date.day)
print(begin_date.hour)
```

```python
date_index = begin_date
origin_list = []
while date_index != end_date:
    month = date_index.month
    day = date_index.day
    hour = date_index.hour
    df = Generate_Origin(month, day, hour, path_origin)
    # file = 'weather_data_2024_{}_{}_{}.xlsx'.format(month, day, hour)
    # try:
    #     origin = pd.read_excel(path + file)
    # except FileNotFoundError:
    #     print("File not found:", file)
    #     date_list.append([month, day, hour])
    origin_list.append(df)
    date_index += timedelta(hours=1)
origin_df = pd.concat(origin_list).reset_index(drop=True)
origin_df
```
